```python
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame(data)
```

```python
plt.figure(figsize=(10,6))
plt.hist(df['IMDB Rating'], bins=10, color='skyblue', edgecolor='black')
plt.xlabel('IMDB Rating')
plt.ylabel('Frequency')
plt.title('Distribution of IMDB Ratings')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df['Duration'],df['IMDB Rating'],color='skyblue',alpha=0.7)
plt.xlabel('Duration')
plt.ylabel('IMDB Rating')
plt.title('Scatter plot of IMDB Rating and Movie Durations')
plt.grid(True)
plt.show()
```

```python
genre_counts = data['Genre'].str.split(',').explode().value_counts()

plt.figure(figsize=(8,8))
plt.pie(genre_counts, labels=genre_counts.index, autopct='%1.1f%%', startangle=140, labeldistance=1.1)
plt.title('Distribution of Movie Genres')
plt.axis('equal')
plt.show()
```
